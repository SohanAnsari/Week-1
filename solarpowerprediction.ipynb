```python
import matplotlib.pyplot as plt
```

```python
#plot distribution of power 
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30, kde= True)
plt.title('Distribution of Geneated Power(kW)')
plt.xlabel('Generated Power(kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
important_features = ['temperature_2_m_above_gnd', 'total_cloud_cover_sfc', 'shortwave_radiation_backwards_sfc', 'angle_of_incidence']
for feature in important_features:
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x=df[feature], y=df['generated_power_kw'])
    plt.title(f'{feature} vs Generated Power')
    plt.xlabel(feature)
    plt.ylabel('Generated Power (kW)')
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['generated_power_kw'])
plt.title('Boxplot of Generated Power (kW)')
plt.xlabel('Generated Power (kW)')
plt.show()
```
